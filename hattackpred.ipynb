```python
#thalachh shows a little left skew so we will use IQR
#to find outliers
def iqr(data, feat):
    q1 = np.quantile(data[feat], 0.25)
    q3 = np.quantile(data[feat], 0.75)
    diff = q3 - q1
    
    lower_bound = q1 - (1.5 * diff)
    upper_bound = q3 + (1.5 * diff)
    
    return data[(data[feat] < lower_bound) | (data[feat] > upper_bound)]
```
